```python
import re
import pandas as pd
```

```python
df = pd.read_csv("data/web.csv", header=0, sep="|")

df=df.drop(['direct_visit','organic_visit','paid_search_visit','email_visit'], axis=1)

df['day']=pd.to_datetime(df['day'])
df.set_index(['day'],drop=True, inplace=True)

df = df.groupby('identifier').resample('M').sum()
df=df.drop('identifier', axis=1).reset_index(0)
df['identifier'] = df.identifier.astype(str)
```

```python
start_date = df.index.min().strftime('%b %Y')
end_date = df.index.max().strftime('%b %Y')
```

```python
df = df.groupby('identifier').apply(lambda group: pd.Series({
    'hits': group['hits'].sum(),
    'hits_product_pages': group['product_pages'].sum(),
    'hits_product_pages_percent': (group['product_pages']/group['hits']).mean(),
    'hits_per_visit': (group['hits']/group['visits']).mean().round(2),
    'orders': group['orders'].sum(),
    'orders_per_visit': (group['orders']/group['visits']).mean().round(2),
    'orders_per_month': group.sort_values(by='day')['orders'].tolist(),   
    'amount': group['amount'].sum()
})).reset_index()

#df=df.sort_values(by=['identifier'])
df['icon'] = df.identifier+".png"
```

```python
df = pl.from_pandas(df)
df
```

```python
sel_visits = pl.selectors.ends_with("per_visit")
sel_hits = pl.selectors.starts_with("hits")
sel_orders = pl.selectors.starts_with("orders")

web_data = (
    gt.GT(df)
    .tab_spanner(label="Visits", columns=sel_visits)
    .tab_spanner(label="Hits", columns=sel_hits)
    .tab_spanner(label="Orders", columns=sel_orders)
    .cols_label(
        icon="",
        hits="Total",
        hits_product_pages="Product Pages",
        hits_product_pages_percent=gt.md(""),
        hits_per_visit=gt.md("*hits/visit*"),
        orders="Total",
        orders_per_visit=gt.md("*orders/visit*"),
        orders_per_month="per Month",
        amount="Amount",
        identifier="Shop",
    )
    .fmt_number(
        columns=pl.selectors.ends_with("amount"),
        compact=True,
        pattern="${x}",
        n_sigfig=3,
    )
    .fmt_percent(columns=pl.selectors.ends_with("percent"), decimals=0)
    .tab_style(
        style=gt.style.fill(color="aliceblue"),
        locations=gt.loc.body(columns=sel_visits),
    )
    .tab_style(
        style=gt.style.fill(color="papayawhip"),
        locations=gt.loc.body(columns="amount"),
    )
    .tab_style(
        style=gt.style.text(weight="bold"),
        locations=gt.loc.body(rows=pl.col("identifier") == "3"),
    )
    .fmt_nanoplot("orders_per_month", plot_type="bar")
    .fmt_image("icon", path="icons/")
    .sub_missing(missing_text="")
    .tab_header(title="Great Tables", subtitle=f"{start_date} to {end_date}")
    .tab_source_note(source_note=gt.md("some user generated data *Google Analytics* style"))
)

web_data
```
